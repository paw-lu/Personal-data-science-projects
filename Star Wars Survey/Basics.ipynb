## Overview

```python
import pandas as pd
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
```

```python
star_wars.head()
```

```python
star_wars.columns
```

```python
is_no_ID = pd.isnull(star_wars['RespondentID'])
star_wars = star_wars[is_no_ID == False]
```

```python
star_wars.head()
```

## Cleaning and Mapping Yes/No Columns

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
yes_no = {
    'Yes': True,
    'No': False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].head()
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].head()
```

## Cleaning and Mapping Checkbox Columns

```python
star_wars.head()
```

```python
import numpy as np
movie_map = {np.NaN: False}
for movie in star_wars.loc[1][3:9]:
    movie_map[movie] = True
movie_map

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_map)
```

```python
star_wars.head()
```

```python
rename_cols = {
    'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_1',
    'Unnamed: 4': 'seen_2',
    'Unnamed: 5': 'seen_3',
    'Unnamed: 6': 'seen_4',
    'Unnamed: 7': 'seen_5',
    'Unnamed: 8': 'seen_6'
}
for col in star_wars.columns[3:9]:
    star_wars = star_wars.rename(columns={col:rename_cols[col]})
```

```python
star_wars.head()
```

## Cleaning the Ranking Columns

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars.dtypes
```

```python
new_titles = []
for i in range(0,6):
    new_titles.append('ranking_' + str(i + 1))
```

```python
ranking_map = {}
count = 0
for col in star_wars.columns[9:15]:
    ranking_map[col] = new_titles[count]
    count += 1
```

```python
star_wars = star_wars.rename(columns=ranking_map)
```

```python
star_wars.columns
```

```python
star_wars.loc[:, 'ranking_1':'ranking_6']
```

## Finding the Highest-Ranked Movie

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

star_wars[star_wars.columns[9:15]].mean().plot.bar()
```

Since these are ordered, a lower ranking is better. Ranking_5—The Empire Strikes Back—is the favorite movie. In general, the orignal trilogy is like better than the following prequels.
